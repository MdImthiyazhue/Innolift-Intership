## Dataset 1 — Heart Disease Dataset (Classification)

| Field | Details |
|---|---|
| **Dataset Name** | Heart Disease Dataset (Cleveland Database) |
| **Dataset Source** | [UCI Machine Learning Repository](https://archive.ics.uci.edu/dataset/45/heart+disease) |
| **Problem Statement** | Given a patient's clinical and diagnostic attributes (age, cholesterol, resting blood pressure, chest pain type, exercise-induced angina, etc.), predict whether the patient has heart disease, to support early screening and diagnosis. |
| **ML Problem Type** | Classification (binary: presence vs. absence of heart disease) |
| **Number of Features** | 13 predictive attributes + 1 target column (303 records) |
| **Target Variable** | `target` — 0 = no heart disease, 1 = heart disease present |
| **Reason for Selecting** | A clean, well-documented, widely-used healthcare benchmark dataset. Its manageable size makes it ideal for practicing the full classification workflow while working on a problem with genuine real-world impact. |


```python
# Load the Heart Disease dataset (mirrored as CSV on GitHub, sourced from the UCI ML Repository)
heart_url = "https://raw.githubusercontent.com/sharmaroshan/Heart-UCI-Dataset/master/heart.csv"
heart_df = pd.read_csv(heart_url)

print("Shape:", heart_df.shape)
heart_df.head()
```

```python
heart_df.info()
```

```python
# Target variable distribution
print("Target value counts (0 = no disease, 1 = disease present):")
print(heart_df['target'].value_counts())
print("\nNumber of features (excluding target):", heart_df.shape[1] - 1)
```

## Dataset 2 — House Prices: Advanced Regression Techniques (Regression)

| Field | Details |
|---|---|
| **Dataset Name** | House Prices: Advanced Regression Techniques (Ames Housing Dataset) |
| **Dataset Source** | [Kaggle](https://www.kaggle.com/c/house-prices-advanced-regression-techniques) |
| **Problem Statement** | Predict the final sale price of residential homes in Ames, Iowa, using 79 explanatory variables describing almost every aspect of the property (lot size, quality ratings, year built, garage, basement, neighborhood, etc.). |
| **ML Problem Type** | Regression (continuous target: sale price in dollars) |
| **Number of Features** | 79 explanatory variables + 1 target column (`SalePrice`), 1,460 training records |
| **Target Variable** | `SalePrice` — the final sale price of the house in USD |
| **Reason for Selecting** | One of the most popular beginner-to-intermediate Kaggle competitions, offering rich opportunities to practice feature engineering, missing-data handling, and comparing regression algorithms on a large, realistic mixed-type feature set. |


```python
# Load the House Prices (Ames Housing) dataset (mirrored as CSV on GitHub, sourced from the Kaggle competition)
house_url = "https://raw.githubusercontent.com/hjhuney/Data/master/AmesHousing/train.csv"
house_df = pd.read_csv(house_url)

print("Shape:", house_df.shape)
house_df.head()
```

```python
house_df.info(verbose=False)  # summarized info given the large number of columns
```

```python
# Target variable summary
print("SalePrice summary statistics:")
print(house_df['SalePrice'].describe())
print("\nNumber of features (excluding Id and target):", house_df.shape[1] - 2)
```

## Dataset 3 — Mall Customer Segmentation Data (Clustering)

| Field | Details |
|---|---|
| **Dataset Name** | Mall Customer Segmentation Data |
| **Dataset Source** | [Kaggle](https://www.kaggle.com/datasets/vjchoudhary7/customer-segmentation-tutorial-in-python) |
| **Problem Statement** | Segment mall customers into distinct groups based on their annual income and spending behavior, so a business can design targeted marketing campaigns for each group. |
| **ML Problem Type** | Clustering (unsupervised learning) |
| **Number of Features** | 5 columns: CustomerID, Gender, Age, Annual Income, Spending Score; 200 records |
| **Target Variable** | None — unsupervised problem; Annual Income & Spending Score are used as clustering features rather than a labeled target |
| **Reason for Selecting** | A small, clean dataset that's perfect for a first hands-on introduction to unsupervised learning (K-Means / hierarchical clustering), with an intuitive, easy-to-visualize business use case. |


```python
# Load the Mall Customer Segmentation dataset (mirrored as CSV on GitHub, sourced from Kaggle)
mall_url = "https://raw.githubusercontent.com/erkansirin78/datasets/master/Mall_Customers.csv"
mall_df = pd.read_csv(mall_url)

print("Shape:", mall_df.shape)
mall_df.head()
```

```python
mall_df.info()
```

```python
# No labeled target -- summarize the two features typically used for clustering
print("Annual Income summary:")
print(mall_df['AnnualIncome'].describe())
print("\nSpending Score summary:")
print(mall_df['SpendingScore'].describe())
```

## Final Dataset Selection & Justification

### Selected Dataset: Heart Disease Dataset (Cleveland Database) — UCI Machine Learning Repository

**Justification:**

The Heart Disease dataset was chosen for the Individual Machine Learning Project because it offers the
best balance of real-world relevance, technical depth, and manageable scope. As a classification problem
in healthcare, it directly aligns with an interest in applying AI to meaningful, high-impact domains. The
dataset is small enough (303 records, 13 features) to allow careful, hands-on exploration of every stage
of the ML pipeline — data cleaning, exploratory analysis, feature encoding, model training, and evaluation
— without being overwhelmed by scale, while still being rich enough to support comparing multiple
classification algorithms (Logistic Regression, Random Forest, and Gradient Boosting) and evaluating them
with clinically meaningful metrics such as recall and ROC-AUC, not just accuracy. It is also a
well-documented, peer-reviewed benchmark dataset, which makes it easier to validate results against
established work in the field.
